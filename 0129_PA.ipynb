```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Generate a BA graph with 2000 nodes, d=4, p=1
BA = nx.random_graphs.barabasi_albert_graph(2000, 3)
```

```python
plt.figure(figsize=(12, 8))
nx.draw(BA, with_labels = False, node_size = 5)
plt.show()
```

```python
degree_freq = nx.degree_histogram(BA)
degrees = range(len(degree_freq))
plt.figure(figsize=(12, 8))
plt.loglog(degrees[3:], degree_freq[3:])
plt.xlabel('Degree')
plt.ylabel('Frequency')
```
